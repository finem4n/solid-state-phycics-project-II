```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
```

```python
# zestaw 1

MASS1 = 4
MASS2 = 1
GAMMA1 = 2
GAMMA2 = 3

qa = np.linspace(-np.pi, np.pi, 1000)
omega1 = np.sqrt(((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)+np.sqrt((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)**2-4*MASS1*MASS2*(2*GAMMA1*GAMMA2-GAMMA1*GAMMA2*(2*np.cos(qa)))))/(2*MASS1*MASS2))
omega2 = np.sqrt(((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)-np.sqrt((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)**2-4*MASS1*MASS2*(2*GAMMA1*GAMMA2-GAMMA1*GAMMA2*(2*np.cos(qa)))))/(2*MASS1*MASS2))

qa_sym = symbols('qa')
omega1_sym = sqrt(((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)+sqrt((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)**2-4*MASS1*MASS2*(2*GAMMA1*GAMMA2-GAMMA1*GAMMA2*(2*cos(qa_sym)))))/(2*MASS1*MASS2))
omega2_sym = sqrt(((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)-sqrt((MASS1*GAMMA1+MASS2*GAMMA1+MASS1*GAMMA2+MASS2*GAMMA2)**2-4*MASS1*MASS2*(2*GAMMA1*GAMMA2-GAMMA1*GAMMA2*(2*cos(qa_sym)))))/(2*MASS1*MASS2))
```

```python
# pl = plot(omega1, omega2)
fig, ax = plt.subplots(figsize=(12,8), dpi=400)
ax.set_xlabel(r'$q [a. u.]$')
ax.set_ylabel(r'$\omega(q) [a. u.]$')

ax.plot(qa, omega1, label='Gałąź optyczna ')
ax.plot(qa, omega2, label='Gałąź akustyczna')

plt.xticks((-np.pi, 0, np.pi), (r'$-\dfrac{\pi}{a}$', '0', r'$\dfrac{\pi}{a}$'))

# ax.grid(True, which='both', axis='both', linestyle='-', linewidth=1)
ax.legend()
fig.savefig("out/phonons.jpg", dpi = 1200)
plt.show()
```
